# **GIẢI THÍCH DỮ LIỆU**

***GIẢI THÍCH DỮ LIỆU:***

**1.merchant_id:** ID của thương gia.

**2.merchant_group_id:** ID nhóm thương gia, có thể đại diện cho một nhóm các thương gia liên quan hoặc thuộc cùng một công ty mẹ.

**3.merchant_category_id:** ID của ngành hàng mà thương gia hoạt động, giúp phân loại các thương gia theo lĩnh vực kinh doanh.

**4.mer_subsector_id:** ID của phân ngành hàng, là một phân nhóm nhỏ hơn trong merchant_category_id.

**5-6.mer_numerical_1, 2:** Các đặc trưng số học có thể đại diện cho các số liệu định lượng như doanh thu trung bình hoặc số lượng giao dịch của thương gia.

**7-9.mer_category_1, 2, 4:** Đặc trưng đại diện cho một yếu tố phân loại nào đó.

**10.mer_most_recent_sales_range:** Phạm vi doanh số bán hàng gần đây của thương gia (A là cao nhất, E là thấp nhất).

**11.mer_most_recent_purchases_range:** Phạm vi số lần mua hàng gần đây từ khách hàng của thương gia, cũng được mã hóa tương tự như most_recent_sales_range.

**12-14.mer_avg_sales_lag3, mer_avg_purchases_lag3, mer_active_months_lag3:** Doanh số trung bình, Số lượng mua hàng trung bình, Số tháng thực hiện giao dịch của thương gia trong 3 tháng trước đó.

**15-17.mer_avg_sales_lag6, mer_avg_purchases_lag6, mer_active_months_lag6:** Tương tự trên, nhưng là 6 tháng

**18-20.mer_avg_sales_lag12, mer_avg_purchases_lag12, mer_active_months_lag12:** Tương tự trên, nhưng là 12 tháng

**21-22.mer_city_id, mer_state_id:** ID của thành phố và bang nơi thương gia hoạt động.

**23.trans_authorized_flag:** Xác định xem giao dịch có được ủy quyền (authorized) hay không.

**24.card_id:** ID của thẻ mà khách hàng sử dụng cho giao dịch.

**25-26.trans_city_id, trans_state_id:** ID của thành phố, bang nơi giao dịch được thực hiện.

**27-29.trans_category_1, 2, 3:** Đặc trưng đại diện cho một thuộc tính cụ thể của giao dịch.

**30.trans_installments:** Số lượng kỳ hạn chọn để thanh toán giao dịch này. Nếu giá trị là -1 có thể chỉ rằng đây là một giao dịch không kỳ hạn, 999 có thể là số kỳ hạn chưa xác định.

**31.trans_merchant_category_id:** ID của ngành hàng mà thương gia thuộc về, giúp phân loại các thương gia theo lĩnh vực kinh doanh.

**32.trans_month_lag:** Số tháng trễ giữa thời điểm giao dịch và thời điểm hiện tại (tính bằng số âm).

**33.trans_purchase_amount:** Số tiền đã chi tiêu trong giao dịch này (đã được mã hóa thành khoảng giá trị)

**34.trans_purchase_date:** Ngày mà giao dịch được thực hiện.

**35.trans_subsector_id:** ID của phân ngành hàng mà thương gia thuộc về, là một phân nhóm nhỏ hơn trong merchant_category_id.


**36.first_active_month:** Tháng đầu tiên hoạt động của khách hàng

**37-39.feature_1, 2, 3:** Đặc trưng đại diện cho một thuộc tính cụ thể của giao dịch.

**40.target/OUTPUT**: Mục tiêu của mô hình, giá trị đại diện cho mô hình Recommendation
